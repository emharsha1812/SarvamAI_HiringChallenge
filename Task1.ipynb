```python


def chunk_transcription(result, max_chunk_duration=15):
    segments = result["segments"]
    
    chunks = []
    current_chunk = {"start": None, "end": None, "text": ""}
    chunk_id = 1

    for segment in segments:
        start = segment['start'] * 1000  # convert to milliseconds
        end = segment['end'] * 1000  # convert to milliseconds
        text = segment['text']

        # Check if the current segment can fit within the current chunk
        if current_chunk["start"] is not None and (end - current_chunk["start"]) > max_chunk_duration * 1000:
            # Close the current chunk and add it to the list
            chunks.append({
                "chunk_id": chunk_id,
                "chunk_length": (current_chunk["end"] - current_chunk["start"]) / 1000,
                "text": current_chunk["text"].strip(),
                "start_time": current_chunk["start"] / 1000,
                "end_time": current_chunk["end"] / 1000
            })
            chunk_id += 1
            current_chunk = {"start": start, "end": end, "text": text}
        else:
            # Add segment to current chunk
            if current_chunk["start"] is None:
                current_chunk["start"] = start
            current_chunk["end"] = end
            current_chunk["text"] += " " + text
    
    # Add the last chunk
    if current_chunk["start"] is not None:
        chunks.append({
            "chunk_id": chunk_id,
            "chunk_length": (current_chunk["end"] - current_chunk["start"]) / 1000,
            "text": current_chunk["text"].strip(),
            "start_time": current_chunk["start"] / 1000,
            "end_time": current_chunk["end"] / 1000
        })

    return chunks

```
