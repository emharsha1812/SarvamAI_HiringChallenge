```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Options
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```python


# Initialize the Chrome WebDriver
chrome_options = webdriver.ChromeOptions()
# Add headless option if necessary
# chrome_options.add_argument("--headless")
wd = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# Open the target website
wd.get("https://www.faithcomesbyhearing.com/")

wd.maximize_window()

# Hover over the main navigation link to display the dropdown
main_nav = wd.find_element(By.CSS_SELECTOR, 'a.main-nav-links__item.txt-nav.main-nav-links__item--dropdown-desktop')
actions = ActionChains(wd)
actions.move_to_element(main_nav).perform()

# Wait for the dropdown to be displayed
dropdown_link_xpath = '//a[contains(text(), "Audio Bible Resources")]'
audio_bible_resources_element = WebDriverWait(wd, 10).until(
    EC.visibility_of_element_located((By.XPATH, dropdown_link_xpath))
)

# Extract the href attribute from the link
audio_bible_resources_link = audio_bible_resources_element.get_attribute('href')
print(audio_bible_resources_link)

# Ensure the element is interactable by scrolling it into view
wd.execute_script("arguments[0].scrollIntoView(true);", audio_bible_resources_element)

# Try clicking the element using JavaScript if normal click doesn't work
try:
    audio_bible_resources_element.click()
except:
    wd.execute_script("arguments[0].click();", audio_bible_resources_element)

# Wait for the new page to load (you can specify an element that should be present on the new page)
try:
    
    WebDriverWait(wd, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'h1')))  # Example: wait for an <h1> tag to be present
    print("New page loaded successfully")
    print(wd.title)

    # Optionally, interact with elements on the new page
    bible_mp3_downloader='//a[contains(text(),"Download the Bible")]'
    bible_mp3_element = wd.find_element(By.XPATH, bible_mp3_downloader)
    bible_mp3_element.click()
    try:
        cookie_banner_accept = WebDriverWait(wd, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'a[fs-cc="allow"]'))
        )
        cookie_banner_accept.click()
    except Exception as e:
        print("No cookie banner found or couldn't dismiss it:", e)
    
    language_input = WebDriverWait(wd, 20).until(EC.presence_of_element_located((By.ID, "Search")))
    
    # Wait until the default value 'English' is loaded in the input field
    WebDriverWait(wd, 20).until(lambda wd: language_input.get_attribute('value') == 'English')
    
    language_input.click()
    language_input.send_keys(Keys.CONTROL + 'a')  # Select all text
    language_input.send_keys(Keys.BACKSPACE)
    # Enter the language you want to select
    language_input.send_keys('Marathi')
    dropdown_container = WebDriverWait(wd, 10).until(
        EC.visibility_of_element_located((By.ID, 'language-search-dropdown'))
    )

    # Select the first suggestion from the dropdown
    first_suggestion = dropdown_container.find_element(By.XPATH, './div[1]')
    first_suggestion.click()
    
        # Wait for the version dropdown to be clickable
    version_dropdown_toggle = WebDriverWait(wd, 20).until(
        EC.element_to_be_clickable((By.ID, "w-dropdown-toggle-0"))
    )
    # Click to open the dropdown
    version_dropdown_toggle.click()

    # Wait for the third option in the dropdown to be visible
    third_option = WebDriverWait(wd, 20).until(
        EC.visibility_of_element_located((By.ID, "MARDPIN1DA"))
    )
    # Click the third option
    third_option.click()

        # Wait for the Name field to be present and enter the value
    first_name_input = WebDriverWait(wd, 20).until(
        EC.presence_of_element_located((By.ID, "mp3-name"))
    )
    first_name_input.send_keys("Harshwardhan")  # Replace with the desired first name

    # Wait for the Last Name field to be present and enter the value
    last_name_input = WebDriverWait(wd, 20).until(
        EC.presence_of_element_located((By.ID, "Mp3-Last-Name"))
    )
    last_name_input.send_keys("Fartale")  # Replace with the desired last name

    # Wait for the Email Address field to be present and enter the value
    email_input = WebDriverWait(wd, 20).until(
        EC.presence_of_element_located((By.ID, "mp3-email"))
    )
    email_input.send_keys("harshafartale12@gmail.com")  # Replace with the desired email address

    WebDriverWait(wd, 50).until(
        EC.frame_to_be_available_and_switch_to_it(
            (By.CSS_SELECTOR, "iframe[title='reCAPTCHA']")
        )
    )
    
    # Find the reCAPTCHA checkbox and click it
    captcha_checkbox = wd.find_element(By.ID, "recaptcha-anchor-label")
    captcha_checkbox.click()

    # Switch back to the default content
    wd.switch_to.default_content()
    
    # Wait for the download button to be clickable
    download_button = WebDriverWait(wd, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, 'input.btn-md.btn--send.mb-40.w-button'))
        )

    # Click the download button
    download_button.click()      
except Exception as e:
    print(e)
    
```
